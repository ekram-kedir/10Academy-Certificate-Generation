```python
# generate_certificate.py
import cv2
import os
import numpy as np
from PIL import Image, ImageDraw, ImageFont
import openai
import requests
from io import BytesIO
from dotenv import load_dotenv

# Load environment variables from .env file
# dotenv_path = '/.env'
# load_dotenv(dotenv_path)

# Access the API key
openai.api_key = "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"

# Craft a prompt to generate a certificate background
prompt = """I NEED to test how the tool works with extremely simple prompts. DO NOT add any detail, just use it AS-IS:
I wanted a blank real white paper nothing more just give me the paper with out anything in addition both on the page and in the background and make it white and blank and also again dont add any addition beside the white blank paper 
"""

# Call the OpenAI API to generate the background
response = openai.Image.create(
    model="dall-e-3",  # Use an available engine
    prompt=prompt,
)

# Print the full response to understand its structure
# Extract the generated image URL from the response
image_url = response['data'][0]['url']
print("Generated image URL:", image_url)
# You might need to use additional methods or tools to extract text from the image URL,
# depending on the specifics of your use case.

# Download the image
image_response = requests.get(image_url)
image_bytes = BytesIO(image_response.content)

# Load the image using OpenCV
image = cv2.imdecode(np.frombuffer(image_bytes.read(), np.uint8), -1)

# Convert the OpenCV image to a PIL image
background_image = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

# Add dynamic information to the certificate
full_name = "John Doe"
logo_path = "./image/logo.png"
date = "January 9, 2024"


logo = cv2.imread(logo_path)
background_image.paste(Image.fromarray(logo), (380, 200))

draw = ImageDraw.Draw(background_image)
font = ImageFont.load_default()

draw.text((400, 410), "Certificate of Completion is awarded to", fill=(0, 0, 0), font=font)
draw.text((440, 430), f"{full_name}", fill=(0, 10, 40), font=font)
draw.text((410, 450), "for successfully completing", fill=(0, 0, 0), font=font)
draw.text((310, 470), "Machine Learning Engineering, Data Engineering, and Web Engineering Training", fill=(0, 0, 0), font=font)
draw.text((290, 500),  """Certifies the mastery in Machine Learning,Data Engineering, and Web Engineering through the
successful completion of a challenging program.For details on the skills acquired, please refer to
the extended document attached.""", fill=(0, 0, 0), font=font)

draw.text((380, 600), "Congratulations on this remarkable achievement!", fill=(0, 0, 0), font=font) 

sign_path = "./image/download.png" 
sign = cv2.imread(sign_path)
background_image.paste(Image.fromarray(sign), (417, 732))    

draw.text((600, 860), f"Date: {date}", fill=(0, 0, 0), font=font)

# Convert the PIL image to a NumPy array for use with cv2
certificate_image = np.array(background_image)

# Display the certificate image using cv2
cv2.imshow("Certificate", certificate_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Save the final certificate image
cv2.imwrite("certificate.png", certificate_image)
```
